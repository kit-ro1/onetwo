# Анализ продаж компьютерных игр

Наш заказчик - интернет-магазин, специализирующийся на продаже компьютерных игр. География продаж - весь мир.   
Чтобы определить потенциально прибыльный продукт и спланировать рекламные кампании, нужно найти критерии, влияющие на популярность игр.

**Цель исследования**: выявить факторы, определяющие успешность игры. 
Успешность будем оценивать по количеству проданных копий.

**Данные**: информация до 2016 года собрана из открытых источников, которая содержит данные о компьютерных играх, их продажах в нескольких регионах, оценках пользователей и экспертов.

- определим актуальный период, данные за который будем использовать для дальнейшего изучения;
- определим потенциально прибыльные платформы на основе динамики продаж;

Возможно, это наблюдение поможет нам заполнить пропуски в дате выпуска. Сгруппируем данные по названию; если в группе есть значение года выпуска, пропуски будут заполнены им.

```python
# заполнение пропусков первым значением года в группе по названию игры
df = df.fillna({'year_of_release': df.groupby('name')['year_of_release'].transform('first')})
```

```python
# подсчет пропусков
df.year_of_release.isna().sum()
```

Изначально в датасете было 269 пропусков в дате, нам удалось заполнить 123 из них.

Остальные пропуски в дате выпуска в рамках нашего исследования восстановить не представляется возможным. И мы не можем удалить эти строки, так как они содержат полезную информацию о продажах. Заполним пропуски "заглушкой" и приведем год к целочисленному формату.

```python
# заполнение пропусков
df.year_of_release = df.year_of_release.fillna(1950)

df.year_of_release = df.year_of_release.astype('int')

# подсчет пропусков
df.year_of_release.isna().sum()
```

#### Рейтинг

В датасете 40% записей отмечено рейтингом E ("Everyone"), означающим доступность для всех возрастных груп.
```python
# замена К-А на Е
df.loc[(df.rating == 'K-A'), 'rating']  = 'E'
```

Столбец с рейтингом содержит большое количество пропусков, которое мы можем заполнить только индикатором пропуска. В качестве такого индикатора используем рейтинг RP ("Rating Pending").

```python
# заполнение пропусков
df.rating = df.rating.fillna('RP')

# подсчет пропусков
df.rating.isna().sum()
```

#### Оценки

```python
# вывод значений столбца user_score
df.user_score.value_counts(normalize=True)
```

Как мы уже отметили выше, столбцы с оценками критиков и пользователей содержат большую долю пропусков: 51% и 40% соответственно. Помимо этого 24% значений пользовательских оценок заполнено аббревиатурой *tbd* (to be determined), которое по сути тоже является пропуском.

Заменим пропуски и tbd на заглушку (-5) и преобразуем формат столбца user_score на float64 и столбца critic_score на int.

```python
# замена tbd
df['user_score'] = df['user_score'].replace('tbd', -5)

# заполнение пропусков
df.user_score = df.user_score.fillna(-5)
df.critic_score = df.critic_score.fillna(-5)

# изменение формата
df.user_score = df.user_score.astype('float64')
df.critic_score = df.critic_score.astype('int')

df
```

```python
df.user_score.hist(density=True, bins=15)
```

```python
df.critic_score.hist(density=True, bins=11)
```

Как мы видим, пользовательские оценки сосредоточены в области значений больше 5, с пиком на отметке 8 и резким падением количества оценок больше 9. 

Оценки критиков образуют более симметричный график за счет более плавного убывания количества высоких оценок после пика; хотя здесь также большинство оценок сосредоточено во второй половине шкалы и пик расположен около 80.

### Динамика выпуска игр

Посчитаем количество игр в разные годы.

```python
# построение гистограммы
ax = df[df['year_of_release'] > 1950].year_of_release.hist(bins=37, color='#194350')
ax.set_ylabel('количество игр')
ax.set_title('Динамика выпущенных игр')
```

```python
# вывод количества игр за каждый год
display(df[df['year_of_release'] > 1950].groupby('year_of_release')['name'].count().reset_index().head(19).transpose())
df[df['year_of_release'] > 1950].groupby('year_of_release')['name'].count().reset_index().tail(18).transpose()
```

Как мы видим, пик развития игровой индустрии приходится на 2008-2009 годы. Начальный этап (до 1990) характеризуется небольшим объемом выпуска и почти полным отсутствием роста. Затем за 8 лет, с 1991 по 1998, объем выпуска увеличивается в 9 раз. Через несколько лет, в 2002, происходит самый крупный скачок за год: почти на 100% по сравнению с предыдущим годом. После пары лет небольшого спада мы снова отмечаем значительный рост, темпы которого сохраняются по 2008. Затем следует сокращение выпуска, которое в 2012 году превращается в резкое падение объема почти в 2 раза по сравнению с предыдущим годом. 2014-2015 года отмечены некоторым ростом - возможно, здесь можно говорить о начале нового периода в развитии рынка. 

Однако нами были посчитаны все строки за каждый год, что соответствует количеству релизов. Возможно, взрывной рост, который мы наблюдали, объясняется тем, что игры выпускались в эти годы на максимальном количестве платформ. Тогда нам нужно посчитать, как соотносится количество релизов с количеством уникальных игр.

```python
# функция для подсчета уникальных значений при группировке
nunq = lambda x: x.nunique()

# группировка для подсчета уникальных игр (по названию) и всех релизов на всех платформах
year_group = df.groupby('year_of_release').agg({'name': [nunq, 'count']}).reset_index()
year_group.columns=['year', 'game_count', 'release_count']

# вычисление доли уникальных игр среди всех релизов
year_group['release_share'] = year_group['game_count'] / year_group['release_count']
```

```python
year_group
```

Как мы видим, с течением времени доля уникальных игр постепенно уменьшается: игры выпускаются на большем количестве платформ.

```python
# построение гистограммы
ax = year_group[year_group['year'] > 1950].plot(kind='bar', x='year', y=['game_count', 'release_count'], figsize=(10,5), 
               color={'game_count': '#ED4145', 'release_count': '#194350'})
ax.set_ylabel('количество')
ax.set_title('Игры в продаже: уникальные названия и релизы на платформах')
```

В целом, изменение объемов как для игр, так и для релизов, идет параллельно, сохраняя общие тенденции роста и падения.

1991 - год, когда доля уникальных названий во всех релизах начинает отличаться от единицы - совпадает с началом первого этапа роста, который мы уже отметили выше.

Самое значительное увеличение доли релизов в соотношении релизов к играм приходится на 2002 год, когда произошел самый крупный годовой скачок в истории игр.

Говоря о развитии рынка игр, нельзя не принять во внимание и динамику продаж. Дальше мы сравним выводы, полученные в этом разделе, с тем, как менялись продажи (в млн копий) в течение того же периода.

```python
# подсчет проданных копий для каждого года 
platform_sales = df.pivot_table(index='year_of_release', values='all_sales', aggfunc='sum').reset_index()
platform_sales.tail()
```

```python
# построение гистограммы по играм
ax = year_group[year_group['year'] > 1950].plot(kind='bar', x='year', y=['game_count', 'release_count'], figsize=(10,5), 
               color={'game_count': '#ED4145', 'release_count': '#194350'})
ax.set_ylabel('количество')
ax.set_title('Игры в продаже: уникальные названия и релизы на платформах')

# построение гистограммы по продажам
ax = platform_sales[platform_sales['year_of_release'] > 1950].plot(kind='bar', x='year_of_release', y='all_sales', 
                                                                   figsize=(10,5), color='#11B2AA')
ax.set_ylabel('количество копий, млн')
ax.set_title('Динамика продаж')
```

В целом, общая форма графика продаж повторяет динамику количества игр (и релизов). Здесь мы тоже видим рост до пикового значения в 2008 году и спад продаж после него. 

Начальный период относительной стагнации растянут по 1995 год (вероятно, в это время на небольшое количество игр приходились большие доли рынка). Рост продаж начинается в 1996, который характеризуется приростом более чем на 100% по сравнению с предыдущим годом.

Пиковое значение 2002 года, которое так ярко выражено на графике количества игр, в продажах предваряется значительным ростом в 2001 году (возможно, здесь есть причинно-следственная связь между ростом продаж в одном году и ростом наименований в следующем). 

Резкое падение от показателей 2011 к 2012 году отмечено на обоих графиках.

Последние годы характеризуются спадом продаж.

Судя по тому, что рынок игр развивается стадиями, при принятия решений на будущий год нужно учитывать последние данные, описывающие текущую стадию развития. Пока мы можем примерно обозначить начало такой стадии 2013-2014 годом.

### Платформы

Посчитаем сумму продаж за все время для каждой платформы. Это позволит нам выделить самые успешные платформы.

```python
# вычисление суммарных продаж по платформам
platform_sales = df.groupby('platform')['all_sales'].sum().reset_index().sort_values(by='all_sales', ascending=False)
platform_sales
```

Любопытно, что рейтинг платформ по продажам не совпадает с рейтингом по количеству релизов, который мы формировали при предварительном обзоре данных. Проиллюстрируем разницу графиками:

```python
# вычисление количества релизов по платформам
platform_copies = df.groupby('platform')['name'].count().reset_index().sort_values(by='name', ascending=False)

# построение графика по количеству релизов
ax = platform_copies.plot(kind='bar', x='platform', y='name', figsize=(10,5), legend=False)
ax.set_ylabel('количество релизов')
ax.set_xlabel('платформа')
ax.set_title('Количество релизов по платформам')

# построение графика по продажам
ax = platform_sales.plot(kind='bar', x='platform', y='all_sales', figsize=(10,5), legend=False)
ax.set_ylabel('количество копий, млн')
ax.set_xlabel('платформа')
ax.set_title('Объем продаж по платформам')
```

### Влияние отзывов на продажи

Проследим влияние отзывов на продажи на данных по PlayStation 4. Построим диаграммы рассеяния и посчитаем корреляцию. 

```python
# построение диаграммы для пользовательских отзывов
ax = df_2013[(df_2013['platform']=='ps4') & (df_2013['user_score']>-5)].plot(
    kind='scatter', x='user_score', y='all_sales', alpha=0.5, color='#0c7d77')
ax.set_xlabel('оценка пользователей')
ax.set_ylabel('копии, млн')
ax.set_title('PS4: пользовательские оценки и объем продаж')

# построение диаграммы для отзывов критиков
ax = df_2013[(df_2013['platform']=='ps4') & (df_2013['critic_score']>-5)].plot(
    kind='scatter', x='critic_score', y='all_sales', alpha=0.5, color='#0c7d77')
ax.set_xlabel('оценка критиков')
ax.set_ylabel('копии, млн')
ax.set_title('PS4: оценки критиков и объем продаж')
```

```python
print('Корреляция между продажами и отзывами пользователей:',
      df_2013[(df_2013['platform']=='ps4') & (df_2013['user_score']>-5)]['all_sales'].corr(
          df_2013[(df_2013['platform']=='ps4') & (df_2013['user_score']>-5)]['user_score']))
print('Корреляция между продажами и отзывами критиков:',
      df_2013[(df_2013['platform']=='ps4') & (df_2013['critic_score']>-5)]['all_sales'].corr(
          df_2013[(df_2013['platform']=='ps4') & (df_2013['critic_score']>-5)]['critic_score']))
```

Расчеты показывают умеренную корреляцию между отзывами критиков и продажами - и отсутствие взаимосвязи между продажами и отзывами пользователей.

На графике с оценками пользователей мы видим, что игры, продажи которых преодолели довольно высокую отметку в 3 млн копий, расположены равномерно от низких оценок (от 3,5) до высоких. Основная масса игр сосредоточена на отрезке от 5,5 до 9 баллов и характеризуется продажами до 1 млн копий.

Корреляция между продажами и отзывами критиков на графике отражена в небольшом наклоне вправо основной массы точек, увеличение объемов продаж начинается примерно в районе 60 баллов. 

Сравним полученные выводы с данными по остальным платформам.

```python
# подсчет количества значений для каждой платформы (за исключением пропусков в оценках пользователей)
display(df_2013[df_2013['user_score']>-5]['platform'].value_counts())

# подсчет количества значений для каждой платформы (за исключением пропусков в оценках критиков)
df_2013[df_2013['critic_score']>-5]['platform'].value_counts()
```

Данных по платформам psp, ds и wii слишком мало, поэтому мы исключим их из расчетов коэффициента корреляции.

```python
# подсчет корреляции и формирование датасета
df_corr = pd.DataFrame(columns=['corr_users', 'corr_critics'], 
                       index=['ps4', 'psv', 'ps3', '3ds', 'xone', 'pc', 'x360', 'wiiu'], dtype="float")

for i in ['ps4', 'psv', 'ps3', '3ds', 'xone', 'pc', 'x360', 'wiiu']:
    corr_users = df_2013[(df_2013['platform']==i) & (df_2013['user_score']>-5)]['all_sales'].corr(
        df_2013[(df_2013['platform']==i) & (df_2013['user_score']>-5)]['user_score'])
    df_corr.loc[(df_corr.index==i), 'corr_users'] = corr_users
    
    corr_critics = df_2013[(df_2013['platform']==i) & (df_2013['critic_score']>-5)]['all_sales'].corr(
        df_2013[(df_2013['platform']==i) & (df_2013['critic_score']>-5)]['critic_score'])
    df_corr.loc[(df_corr.index==i), 'corr_critics'] = corr_critics

df_corr    
```

```python
# построение тепловой карты по полученному датасету
my_colors = ['#41c1bb', '#58c9c4', '#70d1cc', '#b8e8e6', '#cff0ee', '#b8e8e6', '#70d1cc', '#58c9c4', '#41c1bb']
my_cmap = ListedColormap(my_colors)

axs = sns.heatmap(df_corr, annot=True,  vmin=-0.5, vmax=0.5, cmap=my_cmap)
plt.ylabel('платформа')
plt.xlabel('тип отзывов')
plt.title('Коффициент корреляции между отзывами и продажами по платформам')
```

Наиболее выражена корреляция между оценками критиков и продажами для платформ PS4 и Xbox One, самых новых платформ. В целом, уровень корреляции по всем платформам выше для оценок критиков, чем для оценок пользователей. 

Корреляция (умеренная) между продажами и отзывами пользователей обнаружена только для консоли Wii U. Выведем данные на графике:

В следующую группу, релизов для которой было выпущено гораздо меньше, входят гонки, файтинг, платформеры, симуляторы, и стратегии. Самый малочисленный жанр - головоломки.

Возможно, большой отрыв жанра экшен по релизам связан с тем, что в этом жанре игры выпускаются обычно для максимального количества доступных платформ. Посчитаем количество платформ, для которых выпускаются игры в каждом жанре.

```python
# подсчет упоминаний платформ в каждом жанре
df_2013.groupby('genre')['platform'].nunique().sort_values(ascending=False)
```

Верхнюю часть списка с максимальным количеством платформ (11) занимают экшен и несколько игр из второй по популярности группы; в самом конце головоломки.

Мы сделали простой подсчет упоминаний, в который могли войти нестандартные релизы и который не дает информации о том, насколько характерным и массовым является выпуск игр для конкретного жанра на конкретной платформе. Посчитаем теперь распределение релизов по платформам внутри каждого жанра.

```python
# подсчет количества релизов по плафтормам и жанрам
genre_release = df_2013.pivot_table(index='platform', columns='genre', values='name', aggfunc='count')
genre_release = genre_release.fillna(0)
genre_release
```

```python
# вычисление доли платформ внутри жанра
genre_release_share = (100 * genre_release / genre_release.sum()).round(1)
genre_release_share = genre_release_share.transpose()

# построение тепловой карты
axs = sns.heatmap(genre_release_share, cmap="mako", annot = True)
plt.ylabel('жанр')
plt.xlabel('платформа')
plt.title('Доля платформ в релизах по жанрам')
```

Наибольшую концентрацию релизов на одной платформе (53% на 3DS) демонстрирует жанр головоломки, самый малочисленный жанр.

Также относительно высокую концентрацию показывают другие жанры из третьей группы по количеству релизов: стратегии (36% на ПК), симуляторы (34% на ПК), файтинг (30% на PS3).

При этом один жанр из группы более популярных тоже отмечен довольно высокой концентрацией на одной платформе: 35% релизов в жанре приключений были выпущены для PSV. Самая высокая доля одной платформы в остальных популярных жанрах держится на уровне 20-27%. 

Релизы в жанре экшен действительно распределены довольно равномерно, самыми крупными платформами оказались PS4 (19%), PS3 (15%) и 3DS (15%). Но, как нам кажется, такое равномерное распределение лишь отчасти объясняет количественное преимущество жанра. Как мы видим, примерно тот же характер равномерности присущ и для другой категории жанров (misc).

```python
# построение графика по продажам по жанрам
ax = df_2013.groupby('genre')['all_sales'].sum().sort_values(ascending=True).plot(kind='barh', figsize=(10,5))
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel('жанры')
ax.set_xlabel('количество копий, млн')
ax.set_title('Объемы продаж по жанрам за 2013-2016')
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC

```python
# подготовка выборок
df_user_nonull = df_2013.dropna(subset=['user_score'])
sample_pc = df_user_nonull[df_user_nonull['platform']=='pc']['user_score']
sample_x = df_user_nonull[df_user_nonull['platform']=='xone']['user_score']
print('Среднее пользовательского рейтинга платформы Xbox One:', round(sample_x.mean(), 2))
print('Среднее пользовательского рейтинга платформы ПК', round(sample_pc.mean(), 2))
```

Мы видим, что среднее по оценкам различается, но поскольку наши расчеты произведены на данных выборок, а не генеральных совокупностей, этого пока не достаточно, чтобы говорить о существующем различии.

Применим t-тест для проверки. Пользовательские оценки платформ Xbox One и PC - две независимые совокупности, используем метод ttest_ind. Поскольку выборки отличаются по размеру укажем equal_var равным False.

Нулевая гипотеза $H_{0}$ формулируется о равенстве средних. Наша нулевая гипотеза состоит в том, что средние пользовательские рейтинги платформ Xbox One и PC равны.

Соответственно, альтернативная гипотеза заключается в том, что средние различаются.

```python
# критический уровень статистической значимости
alpha = .05 

# t-тест 
results = st.ttest_ind(sample_pc, sample_x, equal_var=False)

print('p-значение: ', results.pvalue) 

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученный уровень значимости довольно высок, больше критического уровня. Вероятность увидеть в выборке полученные значения относительно высока, что не дает основания отвергнуть гипотезу о том, что средние пользовательские оценки Xbox One и PC равны.

### Средние пользовательские рейтинги жанров Action и Sports 

```python
# подготовка выборок
sample_1 = df_user_nonull[df_user_nonull['genre']=='action']['user_score']
sample_2 = df_user_nonull[df_user_nonull['genre']=='sports']['user_score']
```

```python
print('Среднее пользовательского рейтинга жанра экшен:', round(sample_1.mean(), 2))
print('Среднее пользовательского рейтинга спортивных игр', round(sample_2.mean(), 2))
```

Отличаются ли средние пользовательские рейтинги самых популярных жанров action и sports? Как и раньше, применим t-тест для проверки и используем метод ttest_ind. Поскольку выборки отличаются по размеру укажем equal_var равным False.

$H_{0}$: средние пользовательские рейтинги равны.

Альтернативная гипотеза: средние различаются.

```python
# критический уровень статистической значимости
alpha = .05 

# t-тест 
results = st.ttest_ind(sample_1, sample_2, equal_var=False)

print('p-значение: ', results.pvalue) 

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Как мы видим, получившееся p-значение значительно меньше критического уровня значимости, следовательно, нулевая гипотеза о равенстве средних, скорее всего, не верна.

### Вывод

Мы провели два t-теста, результаты которых: 

1) не смогли опровергнуть гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC равны; 

2) не подтвердили гипотезу о равенстве средних пользовательских оценок для жанров экшен и спортивных игр.
