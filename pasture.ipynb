```python
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
yield_by_season = pd.read_csv('yield_by_season.csv', index_col=0)
work_by_season = pd.read_csv('work_by_season.csv', index_col=0)
fields = pd.read_csv('fields.csv', index_col=0)
nutritional = = pd.read_csv('nutritional.csv', index_col=0)
```

```python
all_harvest = {'картофель': 0, 'рис': 0, 'рыба': 0, 'утка': 0, 'вишня': 0}
all_work = 0
for i in range(1,25):
    print('Сезон', i)
    season_work = 0
    for c in all_harvest:
        culture = c
        work_hours = (work_by_season.loc[work_by_season.name==c, str(i)].values[0]
                        *fields.loc[fields.name==c, 'area_ha'].values[0])
        season_work += work_hours
    print('Работ на', season_work ,'часов')
    all_work += season_work
    
    print('Урожай:')
    for c in all_harvest:
        culture = c 
        harvested_kg = (yield_by_season.loc[yield_by_season.name==c, str(i)].values[0]
                        *fields.loc[fields.name==c, 'area_ha'].values[0])
        all_harvest[culture] += harvested_kg
        if harvested_kg > 0:
            print(culture, harvested_kg, 'кг')  
    
    print()
    
print('Всего', all_harvest)    
print('Всего часов',all_work) 
```

```python
protein = 0
fat = 0 
carbohydrates = 0

for c,v in all_harvest.items():
    protein += v*(nutritional.loc[nutritional.name==c, 'protein'].values[0]/100)
    fat += v*(nutritional.loc[nutritional.name==c, 'fat'].values[0]/100)   
    carbohydrates += v*(nutritional.loc[nutritional.name==c, 'carbohydrates'].values[0]/100) 
print('б', protein, 'ж', fat, 'у', carbohydrates)    
```
